```python
#Now let us consider the company column. There are many number of companys available. we can categorize our data based on these companies. 
data['Company'].value_counts()
```

```python
#We can see that from samsung onwards, all catgories have less number of data points. If we do onehot encoding this we have 19 columns.
#so instead of having these low data point avaiable categories we combined all these categories in to 'other' category using a new function
def add_company(input):
    if input=='Samsung' or input=='Mediacom'or input=='Razer'or input=='Microsoft'or input=='Vero'or input=='Xiaomi'or input=='Chuwi'or input=='Fujitsu'or input=='Google'or input=='LG'or input=='Huawei':
        return 'Other'
    else:
        return input
```

```python
data['Company']=data['Company'].apply(add_company)
```

```python
#there are many. we can grab wehther it is touchscreen or IPS and fill those detail in new 2 columns and drop this screenReolution col
data['Touchscreen']=data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0 )
data['IPS']=data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0 )
```

```python
#Now consider CPU
data['Cpu'].value_counts()
```

```python
# first we drop the speed details for that below line
data['Cpu_name']=data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data['Cpu_name'].value_counts()
```

```python
# there are so many categories. we caan simplify this only considering cpu names without considering there speeds. like intel core i3, intel core i5,intel core i7, AMD, Samsung
# for that we need a function
def set_processor(name):
    if name=='Intel Core i5' or name=='Intel Core i7' or name=='Intel Core i3':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'
```

```python
#new column as cpu name with above fn. later we can remove Cpu col
data['Cpu_name']=data['Cpu_name'].apply(set_processor)
```

```python
#now check the value count in cpu _name column
data['Cpu_name'].value_counts()
```

```python
data.head()
```

```python
#Now consider GPU column
len(data['Gpu'].value_counts())
```

```python
#110 is too much for onehot encoding. let check on possible way to reduce
data['Gpu'].value_counts()
```

```python
#we can catogrize them using first name only
data['Gpu_name']=data['Gpu'].apply(lambda x:" ".join(x.split()[0:1]))
```

```python
data['Gpu_name'].value_counts()
```

```python
# we can see only one data point from ARM GPU . so we can delete that row ompletely since not much effect from that GPU type
data.shape
```

```python
#delete ARM GPU name inclduing data row
data=data[data['Gpu_name']!='ARM']
```

```python
data.shape
```

```python
data.head(2)
```

```python
#now consider Opsys. 
data['OpSys'].value_counts()
```

```python
#we can combine windows into one category. and add less categories into 'other' category
def set_os(input):
    if input=='Windows 10' or input=='Windows 7' or input=='Windows 10 S':
        return 'Windows'
    elif input=='macOS' or input=='Mac OS X':
        return 'Mac'
    elif input=='Linux':
        return 'Linux'
    else:
        return 'Other'
```

```python
data['OpSys']=data['OpSys'].apply(set_os)# do the operation on the same opSys col
```

```python
data['OpSys'].value_counts()
```

```python
#now lets delete that mentioned columns
data=data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'])
```

```python
data.shape
```

```python
data.head()
```

```python
#Now we need to covert all these column to integer or floating format to perform the regression. so we use oneHot encoding for that 
data=pd.get_dummies(data)
```

```python
data.head()
```

```python
##Now you can see many columns with one hot encoding. there are many columnts now. . check the shape as well
data.shape
```

```python
#####Now lets state the modell building. now the data set is in the format that ML algos can understand
#we know that this data set is related to the regression problem. because we consider a price. we should calcuale it using the features
#So this is not a categorical problem
#Now we should find what are the best hyper parameters to run the model etc
```

```python
#take only the features without price column
x=data.drop('Price_euros',axis=1)
```

```python
x.head()
```

```python
#consider price col as y
y=data['Price_euros']
```

```python
##Now we should install sk learn library to run the regression model
!pip install sklearn
```

```python
!pip install scikit-learn
```

```python
#now create train test data set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
## there are meny regression possible methods some of them are linear regression, lasso regression, decision tree, randomforest .
#so lets check on each algo on the data set and find which one is the best. for that we built a fuction
def model_accuracy(model):
    model.fit(x_train,y_train)
    accuracy=model.score(x_test,y_test)
    print(str(model)+'---->'+str(accuracy))
```

```python
###First run the linear regression model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_accuracy(lr)
```

```python
###now check lass regression
from sklearn.linear_model import Lasso
lasso=Lasso()
model_accuracy(lasso)
```

```python
##now use decision tree
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_accuracy(dt)
```

```python
##now random forest
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_accuracy(rf)
```

```python
### we can see that random forest model gives a good result. So we select random forest this onwards
```

```python
#we did everything with default hyper parameter. we can tune it to improve the accuracy
#to find the best value we can use grid search.
#parameters in random forest: check the random forest documentation. 
from sklearn.model_selection import GridSearchCV

parameters={'n_estimators':[10,50,100],'criterion':['squared_error','absolute_error','poisson']}
grid_obj=GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit=grid_obj.fit(x_train,y_train)

best_model=grid_fit.best_estimator_
```

```python
best_model
```

```python
## it says criterion should absolute error to get best results. so now lets try the best model accuracy
best_model.score(x_test,y_test)
```

```python
##accuracy has bit improved
```

```python
#Import libraries
from tensorflow.keras.utils import img_to_array,load_img
imag=load_img('elephant.jpg')
imag.show()
```

```python
x=img_to_array(imag)
x.shape
```

```python
x=x.reshape((1,)+x.shape)#change shape to make the rank4 
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#create a object with required augmentation features
datagen=ImageDataGenerator(rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,rescale=1/255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
#Now we apply that augmentation object on our elephant image and create augmentated images. we create 10 such images. we set batch size to 1 since we have only 1 elephant image
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir='augmented_images',save_prefix='elephant',save_format='jpg'):
    i +=1
    if i>=10:
        break
```

This project is a basic RNN example for next word prediction. 

```python
import numpy as np
import tensorflow as tf
```

```python
# import the text data into a variable
with open('Australia_History.txt', 'r', encoding='utf-8') as file:
    australiaHistory= file.read()
```

```python
australiaHistory
```

```python
#To convert the text into numbers, we can use the Tokenizer class from Keras. This will create a mapping of each unique word to a unique integer.
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts([australiaHistory])
```

```python
tokenizer.get_config()
```

```python
#check the word index mapping
tokenizer.word_index
```

```python
#to see the sentences in the given text
input_sequences = []
text=australiaHistory.replace('\n', '.')
for sentence in text.split('.'):
    sentence = sentence.strip()# to avoid any leading or trailing whitespace and removing any empty sentences
    if sentence:
        print (sentence)
        token_list = tokenizer.texts_to_sequences([sentence])[0]
        print (token_list)
```

```python
#create input vs output format for above sentneces. 
input_sequences = []
text=australiaHistory.replace('\n', '.')
for sentence in text.split('.'):
    sentence = sentence.strip()# to avoid any leading or trailing whitespace and removing any empty sentences
    if sentence:
        token_list = tokenizer.texts_to_sequences([sentence])[0]
        for i in range(1, len(token_list)):
            n_gram_sequence = token_list[:i+1]
            input_sequences.append(n_gram_sequence)
```

```python
input_sequences[2]#check on first 5 sequences
```

```python
#to arrange the all sequences into same length, we can use the pad_sequences function from Keras. This will add padding to the sequences so that they all have the same length.
#first we need to find the higest input sequence length
max_sequence_len = max([len(x) for x in input_sequences])
max_sequence_len
```

```python
#Conduct padding
input_sequences = tf.keras.preprocessing.sequence.pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')#padding at the beginning of the sequence
input_sequences[:5]#check on first 5 sequences after padding
```

```python
#create x and y for model training
X = input_sequences[:,:-1] #all the columns except the last one will be our input
y = input_sequences[:,-1]#the last column will be our output
```

```python
X[0]
```

```python
y
```

```python
#curently y has integer values representing the next word in the sequence. We need to convert these integer values into one-hot encoded vectors so that we can use them for training our model.
#first we have to dientify the no of different words in the text, which will be the size of our vocabulary. We can get this from the tokenizer.word_index
vocab_size = len(tokenizer.word_index) + 1 #we add 1 because the word index starts from 1, and we need to account for the 0 index which is used for padding
Y = np.array(tf.keras.utils.to_categorical(y, num_classes=vocab_size))
```

```python
Y[10]#check with 10th value of Y to see the one hot encoding
```

```python
#Now we can create the model using an embedding layer and an LSTM layer. The embedding layer will convert the integer values in X into dense vectors of fixed size, and the LSTM layer will learn the patterns in the sequences to predict the next word.
model = tf.keras.models.Sequential()
model.add(tf.keras.Input(shape=(max_sequence_len - 1,)))#input layer with shape as the length of the input sequences (which is max_sequence_len - 1 because we have removed the last word for the output)
model.add(tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=80))#embedding layer with input dimension as the size of the vocabulary and output dimension as 80 (this is a hyperparameter that we can tune)
model.add(tf.keras.layers.LSTM(100))#LSTM layer with 100 units
model.add(tf.keras.layers.Dense(vocab_size, activation='softmax'))#output layer with softmax activation(becuase this a multi class classification) to get probabilities for each word in the vocabulary
```

```python
model.summary()
```

```python
#compile the model with categorical crossentropy loss and adam optimizer
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#train the model using 100 epochs
history=model.fit(X, Y, epochs=100)
```

```python
#now we can predict using this model.Let us consider below input seqeunce
input_text="the first people to arrive in Australia" 
```

```python
#arrange the tokenlist for given input text
token_list_for_prediction = tokenizer.texts_to_sequences([input_text])[0]
print (token_list_for_prediction)  
```

```python
#add padding to the token list list to make it same length as the input sequences used for training
token_list_for_prediction = tf.keras.preprocessing.sequence.pad_sequences([token_list_for_prediction], maxlen=max_sequence_len-1, padding='pre')
token_list_for_prediction
```

```python
#take the next word prediction from the model
predicted_value=np.argmax(model.predict(token_list_for_prediction))#the model will return a probability distribution over the vocabulary, and we can use np.argmax to get the index of the word with the highest probability, which will be our predicted next word.
print (predicted_value)
```

```python
#check the predicted word from the index
for word, index in tokenizer.word_index.items():
    if index == predicted_value:
        print (word)
        break
```

```python
"""
We can use the same approch to predict multiple words by adding the predicted word to the input sequence and predicting again. 
This way we can generate a sequence of words based on the given input text.
"""
input_text="on April 25 "
for _ in range(10):#predict next 10 words
    token_list_for_prediction = tokenizer.texts_to_sequences([input_text])[0]
    token_list_for_prediction = tf.keras.preprocessing.sequence.pad_sequences([token_list_for_prediction], maxlen=max_sequence_len-1, padding='pre')
    predicted_value=np.argmax(model.predict(token_list_for_prediction))
    for word, index in tokenizer.word_index.items():
        if index == predicted_value:
            input_text += ' ' + word
            break
print (input_text)
```

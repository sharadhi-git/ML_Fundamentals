```python
#import libraries
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder #encode other format of data into numerical format
from sklearn.preprocessing import MinMaxScaler #make all data column into a std scale 
```

```python
#load data set
data=pd.read_csv("sonar.csv",header=None)#since there is no headings in each column, make header=none
data=data.values
```

```python
data[:1]#check first row
```

```python
#Now we devide the data set into x values and y values
X=data[:,0:60].astype(float)
Y=data[:,60]#last column of the dataset
```

```python
#data preprocessing
#Currently y data is in the text format. Hence encode Y data into a numerical format.
encoder=LabelEncoder()
encoder.fit(Y)
Y=encoder.transform(Y)
```

```python
#data preprocessing
#bring all x data columns ino same scale
scaler=MinMaxScaler()
X=scaler.fit_transform(X)
```

```python
#Divide the x and y dataset into training and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#create the model. 
# This is a classification problem. Therefore we create a neural network with 4 layers. 
model=tf.keras.Sequential([
    tf.keras.layers.Dense(30,activation='relu',input_shape=(60,)),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dense(30,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
#complie the model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
            metrics=["accuracy"])
```

```python
#train model
epoch_number=100
history=model.fit(X_train,Y_train,epochs=epoch_number)
```

```python
#We can see that model has achieved a accuracy of 1 for the training dataset. 
# Let us see what is the testing accuracy as well. it is the most important parameter. it is around 0.88 only. This is happen becuase model is overfitted to the training dataset.
model.evaluate(X_test,Y_test)
```

```python
#create the new model with dropout regularization. 
# This is a classification problem. Therefore we create a neural network with 4 layers. 
model2=tf.keras.Sequential([
    tf.keras.layers.Dense(30,activation='relu',input_shape=(60,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(30,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
#complie the new model
model2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
            optimizer=tf.keras.optimizers.Adam(learning_rate=0.005),
            metrics=["accuracy"])
```

```python
#train new model. now it train while dropping out neurons.
epoch_number=100
history2=model2.fit(X_train,Y_train,epochs=epoch_number)
```

```python
model2.evaluate(X_test,Y_test)# now accuracy gets better for the training set. However this changes when we run the model. beucase dropout is random, so model outcome is changed
```

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model ## to visualize the architecutre of the model
```

one input layer, one output layer Squential API -

```python
model1=tf.keras.Sequential([
    tf.keras.layers.Dense(64,activation='relu',input_shape=(2,)),
    tf.keras.layers.Dense(16,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
    ])
```

```python
plot_model(model1,show_shapes=True)
```

One input one output Functional API

```python
input=tf.keras.layers.Input(shape=(2,))
hidden1=tf.keras.layers.Dense(64,activation='relu')(input)
hidden2=tf.keras.layers.Dense(16,activation='relu')(hidden1)
output=tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)
```

```python
model2=tf.keras.Model(inputs=[input],outputs=[output])
```

```python
plot_model(model2,show_shapes=True)
```

two inouts one output: we cannot use a seqeuntial API for this task. we have to use a functional API . first start with two input layers, then hidden layers, and later concatenate and provide the last layer

```python
input1=tf.keras.layers.Input(shape=(2,))
norm1=tf.keras.layers.Normalization()(input1)
hidden1=tf.keras.layers.Dense(16,activation="relu")(norm1)

input2=tf.keras.layers.Input(shape=(5,))
hidden2=tf.keras.layers.Dense(32,activation="relu")(input2)

concat=tf.keras.layers.concatenate([hidden1,hidden2])
output=tf.keras.layers.Dense(1,activation='sigmoid')(concat)

model3=tf.keras.Model(inputs=[input1,input2],outputs=[output])
```

```python
plot_model(model3,show_shapes=True)
```

single input, multiple output Functional API

```python
input=tf.keras.layers.Input(shape=(2,))
hidden1=tf.keras.layers.Dense(16,activation='relu')(input)
hidden2=tf.keras.layers.Dense(32,activation='relu')(hidden1)

output1=tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)
output2=tf.keras.layers.Dense(3,activation='softmax')(hidden2)

model4=tf.keras.Model(inputs=[input],outputs=[output1,output2])
```

```python
plot_model(model4,show_shapes=True)
```

Multiple input multiple output

```python
input1=tf.keras.layers.Input(shape=(2,))
norm1=tf.keras.layers.Normalization()(input1)
hidden1=tf.keras.layers.Dense(16,activation="relu")(norm1)

input2=tf.keras.layers.Input(shape=(5,))
hidden2=tf.keras.layers.Dense(32,activation="relu")(input2)

concat=tf.keras.layers.concatenate([hidden1,hidden2])

output1=tf.keras.layers.Dense(1,activation='sigmoid')(concat)
output2=tf.keras.layers.Dense(3,activation='softmax')(concat)

model5=tf.keras.Model(inputs=[input1,input2],outputs=[output1,output2])
```

```python
plot_model(model5,show_shapes=True)
```
